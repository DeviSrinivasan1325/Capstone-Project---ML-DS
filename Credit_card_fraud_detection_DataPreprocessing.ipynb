```python
import pandas as pd
```

```python
dataset = pd.read_csv('credit_card_fraud_10k.csv')
dataset
```

```python
# Remove unwanted feature
df=dataset.drop(columns=['transaction_id'])
df
```

```python
# Checking for null or missing value
df.isnull().sum()
```

```python
# Converting categorical data into numeric
df = pd.get_dummies(df, dtype= int, drop_first=True)
df
```
